```python
import matplotlib.pyplot as plt
import numpy as np
from environment import k_BanditGameNS
from agent import SimpleAgent
from utils import experiment
```

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
sample_avg_params = {
    "agent": SimpleAgent,
    "env": k_BanditGameNS,
    "agent_params": {
        "epsilon": 0.1,
        "step_policy": "average",
    },
    "n_runs": 2000,
    "n_step": 10000,
}
sample_avg_actions, sample_avg_rewards = experiment(**sample_avg_params)
```

```python
constant_params =  {
    "agent": SimpleAgent,
    "env": k_BanditGameNS,
    "agent_params": {
        "epsilon": 0.1,
        "step_policy": "constant",
        "step_size": 0.1
    },
    "n_runs": 2000,
    "n_step": 10000,
}
constant_actions, constant_rewards = experiment(**constant_params)
```

```python
plt.xlabel("Steps")
plt.ylabel("Average Reward")
plt.plot(sample_avg_rewards, label="Agent with sample average step policy")
plt.plot(constant_rewards, label=r"Agent with constant step policy with $\alpha = 0.1$")
plt.legend()
```

```python
plt.xlabel("Steps")
plt.ylabel("%Optimal actions")
plt.plot(sample_avg_actions * 100, label="Agent with sample average step policy")
plt.plot(constant_actions * 100, label=r"Agent with constant step policy with $\alpha = 0.1$")
plt.legend()
```
