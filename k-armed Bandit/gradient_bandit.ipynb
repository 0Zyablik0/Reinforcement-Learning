```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from environment import k_BanditGame
from agent import GradientAgent
from utils import experiment
```

```python
params = {
    "agent": GradientAgent,
    "env": k_BanditGame,
    "agent_params": {
        "baseline": True
    },
    "n_runs": 2000,
    "n_step": 1000,
}
with_baseline_1_actions, with_baseline_1_reward  = experiment(**params)
```

```python
params = {
    "agent": GradientAgent,
    "env": k_BanditGame,
    "agent_params": {
        "baseline": True,
        "step_size": 0.4,
    },
    "n_runs": 2000,
    "n_step": 1000,
}
with_baseline_2_actions, with_baseline_2_reward  = experiment(**params)
```

```python
params = {
    "agent": GradientAgent,
    "env": k_BanditGame,
    "agent_params": {
        "baseline": False
    },
    "n_runs": 2000,
    "n_step": 1000,
}
without_baseline_1_actions, without_baseline_1_reward  = experiment(**params)
```

```python
params = {
    "agent": GradientAgent,
    "env": k_BanditGame,
    "agent_params": {
        "baseline": False,
        "step_size": 0.4,
    },
    "n_runs": 2000,
    "n_step": 1000,
}
without_baseline_2_actions, without_baseline_2_reward  = experiment(**params)
```

```python
plt.xlabel("Steps")
plt.ylabel("Average Reward")
plt.plot(with_baseline_1_reward, label=r"Gradient Agent with baseline and $\alpha = 0.1$")
plt.plot(without_baseline_1_reward, label=r"Gradient Agent without baseline and $\alpha = 0.1$")
plt.plot(with_baseline_2_reward, label=r"Gradient Agent with baseline and $\alpha = 0.4$")
plt.legend()
```

```python
plt.xlabel("Steps")
plt.ylabel("%Optimal actions")
plt.plot(with_baseline_1_actions, label=r"Gradient Agent with baseline and $\alpha = 0.1$")
plt.plot(with_baseline_2_actions, label=r"Gradient Agent with baseline and $\alpha = 0.4$")
plt.legend()
```
