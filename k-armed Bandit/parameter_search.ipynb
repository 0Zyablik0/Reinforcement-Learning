```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from environment import k_BanditGame
from agent import SimpleAgent, OptimisticSimpleAgent, UCBSimpleAgent, GradientAgent
from utils import experiment
```

```python
params = {
    "agent": SimpleAgent,
    "env": k_BanditGame,
    "agent_params": {
        "epsilon": 0.0,
        "step_policy": "average",
    },
    "n_runs": 1000,
    "n_step": 1000,
}
epsilons = [1/(2**i) for i in range(8)]
eps_results = []
for epsilon in epsilons:
    params["agent_params"]["epsilon"] = epsilon
    greedy_optimal_actions, eps_rewards = experiment(**params)
    eps_results.append(np.mean(eps_rewards))
```

```python
params = {
    "agent": OptimisticSimpleAgent,
    "env": k_BanditGame,
    "agent_params": {
        "epsilon": 0.0,
        "optimistic_value": 0.0,
        "step_policy": "constant",
        "step_size": 0.1,
    },
    "n_runs": 1000,
    "n_step": 1000,
}
optimistic_values = [4/(2**i) for i in range(8)]
ov_results = []
for optimistic_value in optimistic_values:
    params["agent_params"]["optimistic_value"] = optimistic_value
    actions, rewards = experiment(**params)
    ov_results.append(np.mean(rewards))
```

```python
params = {
    "agent": UCBSimpleAgent,
    "env": k_BanditGame,
    "agent_params": {
        "step_policy": "average",
        "ucb_c": 2.0,
    },
    "n_runs": 1000,
    "n_step": 1000,
}
c_values = [4/(2**i) for i in range(8)]
ucb_results = []
for c in c_values:
    params["agent_params"]["ucb_c"] = c
    actions, rewards = experiment(**params)
    ucb_results.append(np.mean(rewards))
```

```python
params = {
    "agent": GradientAgent,
    "env": k_BanditGame,
    "agent_params": {
        "baseline": True,
        "step_size": 0.4,
    },
    "n_runs": 1000,
    "n_step": 1000,
}
step_sizes = [4/(2**i) for i in range(8)]
gradient_results = []
for step_size in step_sizes:
    params["agent_params"]["step_size"] = step_size
    actions, rewards = experiment(**params)
    gradient_results.append(np.mean(rewards))
```

```python
from matplotlib.pyplot import xlabel

x_ticks_places = [4, 2, 1, 1/2, 1/4, 1/8, 1/16, 1/32, 1/64, 1/128]
x_ticks_labels = ["4", "2", "1", r"$\frac{1}{2}$", r"$\frac{1}{4}$", r"$\frac{1}{8}$",
           r"$\frac{1}{16}$", r"$\frac{1}{32}$", r"$\frac{1}{64}$", r"$\frac{1}{128}$"]

plt.title("Parameter Search")
plt.xscale("log")
plt.ylabel("Average Reward over first 1000 steps")
plt.xticks(x_ticks_places, x_ticks_labels)
plt.plot(epsilons, eps_results,
         label=r"\epsilon-greedy algorithm varying $\epsilon$")
plt.plot(optimistic_values, ov_results,
         label=r"optimistic greedy algorithm with $\alpha = 0.1$ varying optimistic values")
plt.plot(c_values, ucb_results, label=r"UCB agent varying c")
plt.plot(step_sizes, gradient_results,
         label=r"Gradient agent varying gradient step size")
plt.legend()
```
